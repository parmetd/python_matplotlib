```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing data

city_data = pd.read_csv('data/city_data.csv')
ride_data = pd.read_csv('data/ride_data.csv')
```

```python
#Checking data feed 1
city_data.head()
```

```python
#Checking data feed 2
ride_data.head()
```

```python
#Merging Data into 1
pyber_data = pd.merge(ride_data, city_data, how='outer', on='city')

pyber_data.head(10)
```

```python
#Checking columns actual names
pyber_data.columns
```

```python
#Cleaning up column names.
pyber_data = pyber_data.rename(columns={
    'city' : 'City',
    'date' : 'Date',
    'fare' : 'Fare',
    'ride_id' : 'Ride ID',
    'driver_count' : 'Driver Count',
    'type' : "Type"
})

pyber_data.head()
```

```python
pyber_city_average = pyber_data.groupby(['City']).mean()

pyber_city_average.head(20)
```

```python
pyber_city_sum = pyber_data.groupby(['City']).sum()

pyber_city_sum.head(20)
```

```python
#bubble plot
#Average Fare ($) Per City
#Total Number of Rides Per City
#Total Number of Drivers Per City
#City Type (Urban, Suburban, Rural)"

bubble_plot_df = pyber_data[['City','Fare','Driver Count', 'Type']]
bubble_plot_df.head()
```

```python
average_ride = round(bubble_plot_df.groupby('City')['Fare'].mean(),2)
total_number_rides = bubble_plot_df.groupby('City')['City'].count()
total_number_drivers = bubble_plot_df.groupby('City')['Driver Count'].count()


```

```python
#% of Total Rides by City Type
type_group = bubble_plot_df.groupby('Type')
count_groupbyrides = type_group['Type'].count()
colors = ["gold", "lightskyblue", 'lightcoral']
explode = (0,0,.1)
piechart_percentage_totalrides = count_groupbyrides.plot(kind = 'pie', colors=colors, legend = False, shadow = True, startangle=165, figsize=(9,9), autopct='%1.2f%%',
                                                         explode =explode, fontsize= 14)

piechart_percentage_totalrides.set_title("% of Total Rides by City Types", fontsize = 20)

#remove the yaxis label
piechart_percentage_totalrides.set_ylabel('')

plt.axis("equal")
plt.savefig("pie_city_type_rides.png")
plt.show()
```

```python
#% of Total Fares by City Type
#Fare data for second pie chart
total_fare = bubble_plot_df.groupby('Type')['Fare'].sum()

colors = ["gold", "lightskyblue", 'lightcoral']
explode = (0,0,.1)
piechart_percentage_totalfares = total_fare.plot(kind = 'pie', colors=colors, legend = False, shadow = True, startangle=165, figsize=(9,9), autopct='%1.2f%%',
                                                         explode =explode, fontsize= 14)

piechart_percentage_totalfares.set_title("% of Total Fares by City Types", fontsize = 20)

#remove the yaxis label
piechart_percentage_totalfares.set_ylabel('')

plt.axis("equal")
plt.savefig("pie_city_type_fares.png")
plt.show()
```

```python
#% of Total Drivers by City Type
total_drivers = bubble_plot_df.groupby('Type')['Driver Count'].sum()

colors = ["gold", "lightskyblue", 'lightcoral']
explode = (0,0,.1)
piechart_percentage_totaldrivers = total_drivers.plot(kind = 'pie', colors=colors, legend = False, shadow = True, startangle=165, figsize=(9,9), autopct='%1.2f%%',
                                                         explode =explode, fontsize= 14)

piechart_percentage_totaldrivers.set_title("% of Total Drivers by City Types", fontsize = 20)

#remove the yaxis label
piechart_percentage_totaldrivers.set_ylabel('')

plt.axis("equal")
plt.savefig("pie_city_type_drivers.png")
plt.show()
```

```python
new_data_df = pd.DataFrame({"Average Fare Per City":average_ride, "Total Rides Per City": total_number_rides, "Total Drivers Per City": total_number_drivers})

new_data_df = new_data_df.reset_index()
new_data_df.head()
```

```python
#Cleaning up column name.
city_data = city_data.rename(columns={
    'city' : 'City',
    })

# City Type (Urban, Suburban, Rural)
all_data_df = pd.merge(new_data_df, city_data, how='left')

#Cleaning up column names.
all_data_df = all_data_df.rename(columns={
    'type' : "Type",
    'driver_count' : "Driver Count"
})


all_data_df.head()
```

```python
# Create Bubble Plot
# Obtain the x and y coordinates for each of the three city types
# Urban data
urban_type_df = all_data_df.loc[all_data_df["Type"] == "Urban",:]
urban_x_axis = urban_type_df["Total Rides Per City"]
urban_y_axis = urban_type_df["Average Fare Per City"]
urban_s = urban_type_df["Driver Count"]
urban = plt.scatter(urban_x_axis, urban_y_axis, marker="o", facecolors="lightcoral", edgecolors="black", s=(urban_s*5),\
                   label="Urban")
# Suburban data
suburban_type_df = all_data_df.loc[all_data_df["Type"] == "Suburban",:]
suburban_x_axis = suburban_type_df["Total Rides Per City"]
suburban_y_axis = suburban_type_df["Average Fare Per City"]
suburban_s = suburban_type_df["Driver Count"]
suburban = plt.scatter(suburban_x_axis, suburban_y_axis, marker="o", facecolors="lightskyblue", edgecolors="black", \
                       s=(suburban_s*5), label="Suburban")
# Rural data
rural_type_df = all_data_df.loc[all_data_df["Type"] == "Rural",:]
rural_x_axis = rural_type_df["Total Rides Per City"]
rural_y_axis = rural_type_df["Average Fare Per City"]
rural_s = rural_type_df["Driver Count"]
rural = plt.scatter(rural_x_axis, rural_y_axis, marker="o", facecolors="gold", edgecolors="black", s=(rural_s*5),\
                    label="Rural")
# Labels
plt.xlabel('Total Number of Rides Per City')
plt.ylabel('Average Fares ($)')
plt.title('Pyber Ride Sharing Data')
plt.yticks((40,30,20))

# Show plot
plt.legend(loc="upper right", title="City Types")
plt.grid()
plt.savefig("PyberRideSharingData.png")
plt.show()
```

```python
#Three Observable Trends.
#1) The Bubble plot shows that the most profitable rides are Rural which also the least drivers (pie chart 3) and the least amount of rides (pie chart 1)
#2) However, Urban rides have both the most volume of revenue (pie chart 2) at 62.72% and the most drivers (pie chart 3) at 86.75% and the most rides (pie chart 1) at 68.42% meaning that their rides are short.
#3) Suburban seems to cross the fields warranting further investigation on a subsequent analysis. Perhaps a different pricing strategy could be examined to determined if suburban rides are more airport based and inquiries into duration of rides would definitey help the picture.
```
